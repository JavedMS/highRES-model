```python
import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import display
```

```python
# Available areas in square km after exclusions
Norway_total_area = 385207          #sq kilometer
areas_data = {
    'Criteria': ['lowCF', '+CORINE', '+WDP', '+nibio', '+tech', '+neigh_low', '+neigh_high'],
    'available_area': [124520, 75687, 40215, 33725, 32972, 31225, 22048],  # after applying exclusions
}

df = pd.DataFrame(areas_data)
# Calculate percentages of the total area of Norway
df['Percentage of Norway'] = (df['available_area'] / Norway_total_area) * 100
```

```python
df
```

```python
# Create bar plot for wind onshore (percentage)
fig, ax = plt.subplots(figsize=(8, 4))
bars = ax.bar(df['Criteria'], df['Percentage of Norway'], color='steelblue')
# Adding values above bars
for bar in bars:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:.2f}%', va='bottom', ha='center', fontsize=10)
# Labels and title
ax.set_xlabel('Exclusion Criteria', fontsize=12)
ax.set_ylabel('Percentage of Norway\'s Total Area', fontsize=12)
ax.set_title('Percentage of Total Area Available for Wind onshore', fontsize=14)

# Hide the right and top spines for a cleaner look
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
# Set the y-axis margin to account for label space
ax.margins(y=0.1)

# Rotate X-axis labels for better readability
plt.xticks(rotation=45)

# Tight layout for saving
plt.tight_layout()
```
